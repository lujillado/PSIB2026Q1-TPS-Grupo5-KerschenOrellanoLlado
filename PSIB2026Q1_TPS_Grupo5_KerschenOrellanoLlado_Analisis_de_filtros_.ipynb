```python
!pip install pynwb lindi scipy numpy matplotlib -q

import pynwb
import lindi
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 110
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.3

local_cache = lindi.LocalCache()

# --- NON (interictal) ---
f_non = lindi.LindiH5pyFile.from_lindi_file(
    "https://lindi.neurosift.org/dandi/dandisets/001044/assets/fa897829-d39a-4ce6-88b4-6f4af4bc3220/nwb.lindi.json",
    local_cache=local_cache)
nwb_non = pynwb.NWBHDF5IO(file=f_non, mode='r').read()
seg_non = nwb_non.acquisition["B6_2023_7_23-non1"]

# --- ICTAL ---
f_ictal = lindi.LindiH5pyFile.from_lindi_file(
    "https://lindi.neurosift.org/dandi/dandisets/001044/assets/b62a0af4-03a8-46d2-a8af-c803d9bb5f86/nwb.lindi.json",
    local_cache=local_cache)
nwb_ictal = pynwb.NWBHDF5IO(file=f_ictal, mode='r').read()
seg_ictal = nwb_ictal.acquisition["B6_2023_7_24-ictal5"]

# Parámetros globales
FS    = int(seg_non.rate)   # 1000 Hz
CH    = 3                   # R_CA3 (índice), el más activo en late stage de B6
ONSET = 308                 # segundos — onset conocido en ictal5

# Cargamos solo el canal que vamos a analizar para no bajar todo
data_non   = seg_non.data[:, CH]    # shape (600000,)
data_ictal = seg_ictal.data[:, CH]  # shape (600000,)

# Etiquetas de canales
LABELS = nwb_non.electrodes["label"].data[:]
print(f"Canal analizado: CH{CH} = {LABELS[CH]}")
print(f"FS = {FS} Hz  |  Duración = {len(data_non)/FS:.0f} s")
print(f"\nContexto del paper:")
print("  Hardware aplicó: pasa-alto 0.1 Hz + notch 50 Hz + pasa-bajo 500 Hz")
print("  Los datos del NWB YA vienen con ese filtrado aplicado")
```

¿Hay DC offset o drift?

```python
dc = np.mean(data_non)
f_psd, psd = signal.welch(data_non, fs=FS, nperseg=4*FS)
pot_total  = np.trapezoid(psd, f_psd)
pot_dc     = np.trapezoid(psd[f_psd < 1], f_psd[f_psd < 1])

print("=== PREGUNTA 1: ¿hace falta filtro pasa-alto en software? ===\n")
print(f"  DC offset (media de la señal): {dc:.4f} µV")
print(f"  Energía en 0–1 Hz:             {pot_dc/pot_total*100:.3f}% del total")
print()
if abs(dc) < 5 and pot_dc/pot_total < 0.01:
    print("  → NO hace falta. El hardware ya eliminó el drift.")
else:
    print("  → SÍ hace falta un pasa-alto adicional.")
```

El pasa-alto en 1Hz también es redundante — el hardware (bandpass 0.1-500Hz) ya eliminó el drift DC, y la energía en 0-1Hz es despreciable (0.026%

```python
# ═══════════════════════════════════════════════════════════════════════
# VISUALIZACIÓN: ¿hace falta un pasa-alto adicional por software?
# Panel A: ¿hay drift/tendencia en el tiempo?
# Panel B: RAW vs con pasa-alto 1Hz aplicado — ¿se nota diferencia?
# Panel C: PSD log-log en bajas frecuencias — ¿sube hacia 0Hz?
# Panel D: distribución de energía por banda
# ═══════════════════════════════════════════════════════════════════════
fig, axes = plt.subplots(2, 2, figsize=(14, 8))

# Aplicar pasa-alto 1Hz para comparar
sos_hp  = signal.butter(4, 1, btype='high', fs=FS, output='sos')
data_hp = signal.sosfiltfilt(sos_hp, data_non)

t = np.arange(len(data_non)) / FS

# --- Panel A: señal cruda + media móvil (drift) ---
ax = axes[0, 0]
ax.plot(t, data_non, lw=0.3, color='steelblue', label='RAW')
win = 5 * FS
moving_mean = np.convolve(data_non, np.ones(win)/win, mode='valid')
t_mm = t[win//2 : win//2 + len(moving_mean)]
ax.plot(t_mm, moving_mean, color='red', lw=2, label='media móvil 5s (drift)')
ax.axhline(dc, color='black', ls='--', lw=1, label=f'DC global={dc:.3f}µV')
ax.set_title(f'A — Señal RAW {LABELS[CH]} (non)\n¿hay drift/tendencia?', fontweight='bold')
ax.set_xlabel('Tiempo (s)'); ax.set_ylabel('Amplitud (µV)')
ax.legend(fontsize=8)

# --- Panel B: RAW vs con pasa-alto, zoom 5s ---
ax = axes[0, 1]
seg = slice(0, 5000)
ax.plot(t[seg], data_non[seg], lw=0.8, color='steelblue', label='RAW', alpha=0.8)
ax.plot(t[seg], data_hp[seg],  lw=0.8, color='orange', label='con pasa-alto 1Hz', alpha=0.8)
ax.set_title('B — RAW vs con pasa-alto 1Hz (zoom 5s)\n¿se ve diferencia?', fontweight='bold')
ax.set_xlabel('Tiempo (s)'); ax.set_ylabel('Amplitud (µV)')
ax.legend(fontsize=8)

# --- Panel C: PSD log-log, bajas frecuencias ---
ax = axes[1, 0]
ax.loglog(f_psd, psd, color='steelblue', lw=1.2)
ax.axvline(1, color='red', ls='--', lw=1.5, label='1 Hz (corte pasa-alto típico)')
ax.set_xlim(0.05, 20)
ax.set_title('C — PSD log-log, bajas frecuencias\n¿sube hacia 0Hz (drift) o es plana?', fontweight='bold')
ax.set_xlabel('Frecuencia (Hz)'); ax.set_ylabel('PSD (µV²/Hz)')
ax.legend(fontsize=8)
ax.grid(True, which='both', alpha=0.3)

# --- Panel D: energía por banda ---
ax = axes[1, 1]
bandas = [(0, 1, '0-1Hz\n(DC/drift)'), (1, 4, '1-4Hz\nδ'), (4, 8, '4-8Hz\nθ'),
          (8, 13, '8-13Hz\nα'), (13, 30, '13-30Hz\nβ')]
pots = []
for lo, hi, _ in bandas:
    m = (f_psd >= lo) & (f_psd < hi)
    pots.append(np.trapezoid(psd[m], f_psd[m]) / pot_total * 100)
colors = ['red'] + ['steelblue'] * (len(bandas) - 1)
ax.bar(range(len(bandas)), pots, color=colors)
ax.set_xticks(range(len(bandas)))
ax.set_xticklabels([b[2] for b in bandas], fontsize=8)
ax.set_ylabel('% energía total')
ax.set_title('D — Distribución de energía por banda\n(rojo = banda que eliminaría el pasa-alto)', fontweight='bold')
for i, p in enumerate(pots):
    ax.text(i, p + 0.3, f'{p:.2f}%', ha='center', fontsize=8)

plt.suptitle(f'¿Hace falta pasa-alto en software? — {LABELS[CH]} (non1)\n'
              f'DC offset={dc:.4f}µV | Energía 0-1Hz={pot_dc/pot_total*100:.4f}% del total → REDUNDANTE',
              fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```


Panel A — la media móvil de 5s (línea roja) se mantiene prácticamente plana alrededor de 0: no hay tendencia ni deriva lenta en la señal.
Panel B — al superponer la señal cruda y la filtrada con pasa-alto 1Hz, son prácticamente indistinguibles (la diferencia tiene std=3.47µV contra std=105.94µV de la señal — menos del 3.3%).
Panel C — en escala log-log, la PSD no muestra el típico crecimiento hacia 1/f cerca de 0Hz que indicaría drift; se mantiene en niveles bajos y estables por debajo de 1Hz.
Panel D — la banda 0-1Hz (en rojo, la que eliminaría el pasa-alto) contiene solo 0.026% de la energía total, comparado con 1.85% en delta — confirmando numéricamente que no hay nada relevante (ni ruido ni señal) que el pasa-alto esté removiendo.
Conclusión: el pasa-alto de 1Hz en preprocesar() es redundante — el hardware ya dejó la señal sin DC offset ni drift de baja frecuencia

¿Hay artefacto de red (50 Hz)?

```python
mask_50  = (f_psd >= 48) & (f_psd <= 52)
mask_vec = ((f_psd >= 44) & (f_psd <= 48)) | ((f_psd >= 52) & (f_psd <= 56))
pot_50   = psd[mask_50].mean()
pot_vec  = psd[mask_vec].mean()
ratio_50 = pot_50 / (pot_vec + 1e-10)

print("=== PREGUNTA 2: ¿hace falta notch 50 Hz en software? ===\n")
print(f"  Potencia media en 50 Hz:         {pot_50:.2f} µV²/Hz")
print(f"  Potencia media en vecinos 44-56: {pot_vec:.2f} µV²/Hz")
print(f"  Ratio pico/vecinos:              {ratio_50:.2f}x")
print()
if ratio_50 > 3:
    print("  → SÍ hay pico de red. El notch HW no fue suficiente.")
else:
    print("  → NO hay pico visible. El notch HW ya lo eliminó.")
```

```python
# ═══════════════════════════════════════════════════════════════════════
# VISUALIZACIÓN: ¿hay pico de red eléctrica a 50Hz en los datos crudos?
# Fila superior: PSD completa (0-200Hz, escala log) — contexto general
# Fila inferior: zoom 40-60Hz, marcando banda "50Hz" vs vecinos
# Análisis sobre el canal CH = R_CA3, fragmentos non e ictal
# ═══════════════════════════════════════════════════════════════════════
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

for col, (data, nombre) in enumerate([(data_non, 'NON (interictal)'),
                                       (data_ictal, 'ICTAL5')]):
    f_psd_c, psd_c = signal.welch(data, fs=FS, nperseg=2048, noverlap=1024)

    mask_50  = (f_psd_c >= 48) & (f_psd_c <= 52)
    mask_vec = ((f_psd_c >= 44) & (f_psd_c <= 48)) | ((f_psd_c >= 52) & (f_psd_c <= 56))
    pot_50   = psd_c[mask_50].mean()
    pot_vec  = psd_c[mask_vec].mean()
    ratio_50 = pot_50 / (pot_vec + 1e-10)

    # --- Fila superior: PSD completa, escala log ---
    ax = axes[0, col]
    ax.semilogy(f_psd_c, psd_c, color='steelblue', lw=1)
    ax.axvline(50, color='red', ls='--', lw=1, alpha=0.7)
    ax.set_xlim(0, 200)
    ax.set_title(f'{LABELS[CH]} — {nombre} — PSD completa (raw)',
                  fontsize=10, fontweight='bold')
    ax.set_xlabel('Frecuencia (Hz)')
    ax.set_ylabel('PSD (µV²/Hz)')

    # --- Fila inferior: zoom 40-60Hz con bandas marcadas ---
    ax = axes[1, col]
    ax.plot(f_psd_c, psd_c, color='steelblue', lw=1.5, marker='o', ms=3)
    ax.axvspan(48, 52, color='red', alpha=0.15, label='banda "50Hz"')
    ax.axvspan(44, 48, color='green', alpha=0.12, label='vecinos')
    ax.axvspan(52, 56, color='green', alpha=0.12)
    ax.axvline(50, color='red', ls='--', lw=1.5)
    ax.set_xlim(40, 60)
    ax.set_title(f'Zoom 40-60Hz | ratio={ratio_50:.2f}x', fontsize=10, fontweight='bold')
    ax.set_xlabel('Frecuencia (Hz)')
    ax.set_ylabel('PSD (µV²/Hz)')
    if col == 0:
        ax.legend(fontsize=8)

plt.suptitle(f'PSD señal RAW (sin filtrado por software) — {LABELS[CH]} — ¿hay pico de red a 50Hz?\n'
              'ratio < 1 → valle en 50Hz → notch de HARDWARE ya efectivo',
              fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

El valle que se observa es la "huella" o firma del notch de hardware. No es ruido eléctrico — es la cicatriz que deja el filtro al haber actuado. Por eso el ratio < 1: la potencia en 48-52Hz quedó por debajo de sus vecinas, justo lo opuesto de lo que se observaría si hubiera contaminación de red sin filtrar ( pico agudo, como los que aparecen en 100Hz y 150Hz).

Los picos agudos en 100Hz y 150Hz —  son el 2º y 3er armónico de 50Hz. El notch de hardware está afinado solo en 50Hz (Q=30, banda angosta 48-52Hz) y no atenúa sus armónicos. Esto es evidencia de que sí hubo contaminación real de red eléctrica en la grabación — el hardware la eliminó parcialmente (solo el fundamental).

Conclusión:
El valle en 50Hz confirma que el notch de hardware se aplicó y funcionó sobre el fundamental.
Agregar otro notch de 50Hz por software sería contraproducente: profundizaría un valle que ya existe, eliminando señal biológica real en esa banda (justo donde cae parte de la banda gamma, relevante para detectar crisis según el paper).
Pero los armónicos en 100Hz y 150Hz siguen presentes — podrían justificar un notch adicional en esas frecuencias si te interesa esa banda específicamente (aunque 100-150Hz ya está fuera del rango típico de análisis LFP del paper, que se centra en 0-100Hz).

Si bien el análisis espectral (filtrosiono) mostró que los armónicos de la red eléctrica en 100Hz y 150Hz representan una fracción considerable de la energía total en varios canales (hasta ~ 60% en algunos casos), agregar notches adicionales en esas frecuencias resultó contraproducente para el Método 1. La disimilitud por autocorrelación normaliza por R_ref(0)·R_test(0) — es decir, por la varianza de cada ventana. Al eliminar los armónicos, la varianza de las ventanas se redujo de forma muy desigual: en ventanas de baja amplitud (post-ictal, t≈380-450s), esos armónicos representaban una fracción dominante de la energía, por lo que R_test(0) cayó hasta 8 veces; en ventanas de alta amplitud (durante la crisis), la caída fue mínima (~ 3%). Esta reducción no uniforme del denominador infló artificialmente el coeficiente de disimilitud en ventanas de baseline/post-ictal (llegando a valores de hasta 127, comparables o mayores que durante la crisis real), rompiendo la calibración del umbral mediana+8·MAD y provocando que M1 "detectara" un bloque continuo de 318 a 424s — muy por encima de la duración real de la crisis (~33s). En definitiva, el filtrado adicional mejoró la calidad espectral de la señal en términos absolutos, pero alteró la relación de escala entre ventanas de la que depende la normalización de M1, degradando su capacidad de discriminación. Por este motivo se mantuvo el preprocesamiento original (notch solo en 50Hz), priorizando la consistencia de la métrica de detección sobre la eliminación completa de artefactos de red que, de todas formas, no afectan significativamente al rango de frecuencias relevante para la discriminación ictal/interictal (delta-gamma, <100Hz).

¿Hay señal útil por encima de 200 Hz?

```python
pot_hi = np.trapezoid(psd[f_psd > 200], f_psd[f_psd > 200])

print("=== PREGUNTA 3: ¿hay información sobre 200 Hz? ===\n")
print(f"  Energía >200 Hz: {pot_hi/pot_total*100:.2f}% del total")
print()
if pot_hi/pot_total > 0.03:
    print("  → SÍ hay energía. Pueden ser HFO (High-Frequency Oscillations),")
    print("    biomarcadores epilépticos. Cortarlos con un pasa-bajo 200 Hz")
    print("    significaría descartar información relevante.")
else:
    print("  → Energía despreciable. El corte en 200 Hz es seguro.")

fig, ax = plt.subplots(figsize=(12, 4))
ax.semilogy(f_psd, psd, lw=1, color='steelblue', label=f'PSD {LABELS[CH]} (non)')
ax.axvline(200, color='red',   lw=1.2, ls='--', label='200 Hz (corte típico SW)')
ax.axvline(50,  color='green', lw=0.8, ls=':',  label='50 Hz (notch HW)')
ax.axvline(1,   color='gray',  lw=0.8, ls=':',  label='1 Hz (pasa-alto típico SW)')
ax.set_xlim(0, 510)
ax.set_xlabel('Frecuencia (Hz)')
ax.set_ylabel('PSD (µV²/Hz)')
ax.set_title(f'PSD completa — {LABELS[CH]} — fragmento NON\n¿dónde está la energía?')
ax.legend(fontsize=9)

# Anotar porcentajes en el gráfico
for lo, hi, nombre, color in [
    (0,   1,   '0–1 Hz\n(drift)',         'gray'),
    (1,   200, '1–200 Hz\n(LFP clásico)', 'steelblue'),
    (200, 500, '>200 Hz\n(HFO)',           'red'),
]:
    m = (f_psd >= lo) & (f_psd < hi)
    p = np.trapezoid(psd[m], f_psd[m])
    ax.text((lo + hi) / 2, psd[f_psd > 0].min() * 3,
            f'{nombre}\n{p/pot_total*100:.1f}%',
            ha='center', fontsize=8, color=color)

plt.tight_layout()
plt.show()
```
